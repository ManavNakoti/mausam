```python
import matplotlib.pyplot as plt
import pandas as pd

# Loading the dataset
data = pd.read_csv('dehradun.csv', parse_dates=['datetime'])

# Ploting the temperature trends
plt.figure(figsize=(11, 5))
plt.plot(data['datetime'], data['tempmax'], label='Max Temp')
plt.plot(data['datetime'], data['tempmin'], label='Min Temp')
plt.plot(data['datetime'], data['temp'], label='Avg Temp')
plt.xlabel('Date')
plt.ylabel('Temperature (C)')
plt.title('Temperature Trends Over Time')
plt.legend()
plt.show()
```

```python
import seaborn as sns

plt.figure(figsize=(10, 6))
sns.histplot(data['temp'], kde=True, bins=30, color='skyblue')
plt.title('Temperature Distribution')
plt.xlabel('Temperature (C)')
plt.ylabel('Frequency')
plt.show()
```

```python
plt.figure(figsize=(10, 6))
sns.scatterplot(x='temp', y='humidity', data=data, color='purple')
plt.title('Temperature vs. Humidity')
plt.xlabel('Temperature (C)')
plt.ylabel('Humidity (%)')
plt.show()
```

```python
data['month'] = data['datetime'].dt.month
plt.figure(figsize=(12, 6))
sns.boxplot(x='month', y='temp', data=data)
plt.title('Monthly Temperature Variation')
plt.xlabel('Month')
plt.ylabel('Temperature (C)')
plt.show()
```

```python
plt.figure(figsize=(18, 14))
sns.heatmap(data.corr(), annot=True, cmap='coolwarm')
plt.title('Feature Correlation Heatmap')
plt.show()
```
